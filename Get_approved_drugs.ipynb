```python
from chembl_webresource_client.new_client import new_client
```

```python
import json
```

```python
import pandas as pd
```

```python
molecule = new_client.molecule
approved_drugs = molecule.filter(max_phase=4).order_by('molecule_properties__mw_freebase')
```

```python
len(approved_drugs)
```

```python
print(approved_drugs[0])
```

```python
new_column_names = ['pref_name', 'chembl_id', 'indication_class', 'mw_freebase', 'canonical_smiles']
df_approved_drugs = pd.DataFrame(columns=new_column_names)
```

```python
df_approved_drugs
```

```python
for drug in approved_drugs:
    try:
        df_approved_drugs.loc[len(df_approved_drugs)] = [drug['pref_name'],drug['molecule_chembl_id'], 
                                                         drug['indication_class'], drug['molecule_properties']['mw_freebase'],
                                                         drug['molecule_structures']['canonical_smiles']]
        print(drug['pref_name'],drug['molecule_chembl_id'], drug['indication_class'],
              drug['molecule_properties']['mw_freebase'],
              drug['molecule_structures']['canonical_smiles'])
    except:
        pass
```

```python
df_approved_drugs
```

```python
# Save the DataFrame to a CSV file
df_approved_drugs.to_csv('./results/approved_drugs.tsv', sep = '\t', index=False)
```
